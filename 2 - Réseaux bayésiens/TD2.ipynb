# CLASSIFICATEUR BAYESIEN NAIF

```python
## Classificateur naif bayésien
fw = 0
fnw = 0
for i in range(len(data)):
    if(data[37].loc[i] == "won"):
        fw += 1
    else :
        fnw +=1
pw = fw/len(data)
pnw = fnw/len(data)
print("Nombre de won :", fw)
print("Nombre de nowon :",fnw)
print(" ")
print("Probabilité de won : ", pw)
print("Probabilité de nowon : ",pnw)
```

```python
col = ["bkblk","bknwy","bkon8","bkona","bkspr","bkxbq","bkxcr","bkxwp","blxwp","bxqsq","cntxt","dsopp","dwipd","hdchk","katri","mulch","qxmsq","r2ar8","reskd","reskr","rimmx","rkxwp","rxmsq","simpl","skach","skewr","skrxp","spcop","stlmt","thrsk","wkcti","wkna8","wknck","wkovl","wkpos","wtoeg","class"]
```

```python
#Fct qui permet de voir toutes les valeurs possibles dans une colonne
def colView(col,df):
    #allrev = []
    allrev = df[col]
    revNoDouble = list(set(allrev))
    return revNoDouble
```

```python
# On va calculer la probabilité de chaque colonne
for i in range(len(data.columns)-1):
    pos = colView(i+1,data)
    valw = [0]*len(pos)
    valnw = [0]*len(pos)
    pws = 1
    pnws = 1
    #On parcours le nombre de possibilité de valeur pour chaque colonne
    for t in  range(len(pos)):
        #On parcours le nombre de ligne dans la dataset
        for j in range(len(data)):
            if data[37].loc[j] == "won":
                if pos[t] == data[i+1].loc[j] :
                    valw[t] += 1
            else :
                if pos[t] == data[i+1].loc[j] :
                    valnw[t] += 1
        #On fait le produit de tout les P(X|Ci) pour trouver P(X|won) et P(X|noWon)
        pws = pws * valw[t]/fw
        pnws = pnws * valnw[t]/fnw
        
        
        print("P(",col[i],"=",pos[t],"|won) =",valw[t],"/",fw, " = ", valw[t]/fw )
        #print("Pour index :", i+1, "il y a ",pos[t], " :", valw[t], "si il WON.")
        #print("La probabilité est de : ", valw[t]/fw)
        print("P(",col[i],"=",pos[t],"|noWon) =",valnw[t],"/",fnw, " = ", valnw[t]/fnw )
        print(" ")
        #print("Pour index :", i+1, "il y a ",pos[t], " :", valnw[t], "si il NOWON.")
        #print("La probabilité est de : ", valnw[t]/fnw)
    print("__________________________________________________________________")
        ##print("La probabilité que : ",i, "soit égale a :" post[t], "sachant ")
```

```python
print("P(X|won) = ",pws)
print("P(X|noWon) = ",pnws)
print(" ")
print("P(X|won)*P(won) = ", pws*pw)
print("P(X|noWon)*P(noWon) = ", pnws*pnw)
```
